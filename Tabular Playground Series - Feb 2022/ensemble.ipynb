# Imports

```python
import os
import numpy as np
import pandas as pd
from scipy import stats

from sklearn.preprocessing import LabelEncoder
```

```python
train = pd.read_csv('../input/tabular-playground-series-feb-2022/train.csv')
test = pd.read_csv('../input/tabular-playground-series-feb-2022/test.csv')
```

# Ensemble

```python
target_name = 'target'
ensemble_csvs = [
    '../input/tps-feb-2022-resnet/submission.csv',
    '../input/tpsfeb22-03-clustering-improves-the-predictions/submission.csv',
    '../input/extra-blender-addition/submission.csv',
    '../input/tabular-playground-series-feb-2022-09/submission_6.scv',
    '../input/tps-feb-22/submission.csv'
]
```

```python
le = LabelEncoder()
le.fit(train[target_name])
```

```python
preds = []

for ensemble_csv in ensemble_csvs:
    df = pd.read_csv(ensemble_csv)
    preds.append(le.transform(df[target_name])) 
    
modes = stats.mode(np.array(preds), axis=0)[0][0]
```

```python
submission = pd.DataFrame({
    'row_id': test.row_id,
    target_name: le.inverse_transform(modes)
})
submission.head()
```

```python
submission.to_csv("submission.csv", index=False)
```
